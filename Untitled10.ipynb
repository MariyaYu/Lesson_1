```python
data1 = pd.DataFrame({'Products': ['Snacks non Veg', 'Hot\nWings 2 pc', 'Hot Wings 4 Pc'], 'Avg. Portio n Weigh\nt': ['g', 75, 155]})
data2 = pd.DataFrame({'Products': ['Hot\nWings 2 pc', 'Hot Wings 4 Pc', 'Pepsi Pet'], 'Total Energy': [267.1, 534.3, 143.3]})
result = pd.merge(data1, data2, on='Products', how='inner')
```
