```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import scipy.stats
from scipy.stats import shapiro
from scipy.stats import anderson
from scipy.stats import kstest
from scipy.stats import mannwhitneyu
data = pd.read_csv("food.csv")
df = pd.DataFrame(data)
df
```

```python
df.info()
```

```python
# пропуски отсутствуют. Также визуально кажется, что выбросов и дубликатов нет.
```

```python
df.describe()
```

```python
sns.histplot(df['Data.Water'])
```

```python
result = scipy.stats.shapiro(df['Data.Water'])
if result.pvalue > 0.05:
    print("Распределение нормальное", result.pvalue)
else:
    print("Распределение ненормальное", result.pvalue)
```

```python
result = scipy.stats.anderson(df['Data.Water'])
if result.statistic < result.critical_values[2] and result.significance_level[2] > 0.05:
    print("Распределение нормальное", result.significance_level[2])
else:
    print("Распределение ненормальное", result.significance_level[2])
```

```python
result = scipy.stats.kstest(df['Data.Water'], 'norm')
if result.pvalue > 0.05:
    print("Распределение нормальное", result.pvalue)
else:
    print("Распределение ненормальное", result.pvalue)
```
