```python
data = {'Дата': pd.date_range(start='1985-01-01', periods=10, freq='D'), 'Значение': [195000, 175200, 103800, 78400, 180200, 60800, 70600, 55200, 120800, 167600]}
df = pd.DataFrame(data)
```

```python
df
```

```python
rolling_window = df['Значение'].rolling(window=3)
```

```python
# находим скользящее среднее
```

```python
rolling_mean = rolling_window.mean()
```

```python
df['Скользящее_среднее'] = rolling_mean
```

```python
df['Скользящее_среднее']
```

```python
# находим скользящую сумму
```

```python
rolling_sum = rolling_window.sum()
```

```python
df['Скользящая_сумма'] = rolling_sum
```

```python
df['Скользящая_сумма']
```

```python
# находим скользящую дисперсию
```

```python
df['Скользящая_дисперсия'] = rolling_std
```

```python
## доли рыночной стоимости компаний из этого сектора
data = pd.read_excel('Nasdaq.xlsx')
df = pd.DataFrame(data)
df
# Выбор компаний с сектором Technology
tech_companies = df[df['Sector'] == 'Technology']
# Расчет доли рыночной стоимости компаний Technology
tech_market_cap_ratio = tech_companies['Market Cap'].sum() / df['Market Cap'].sum()
# Расчет доли рыночной стоимости компаний остальных секторов
other_companies = df[df['Sector'] != 'Technology']
other_market_cap_ratio = other_companies['Market Cap'].sum() / df['Market Cap'].sum()
# Визуализация результатов
labels = ['Technology', 'Other Sectors']
ratios = [tech_market_cap_ratio, other_market_cap_ratio]
plt.bar(labels, ratios)
plt.xlabel('Sectors')
plt.ylabel('Market Cap Ratio')
plt.title('Market Cap Ratio of Technology vs Other Sectors')
plt.show()
```
