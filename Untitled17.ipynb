```python
## доли рыночной стоимости компаний из этого сектора
data = pd.read_excel('Nasdaq.xlsx')
df = pd.DataFrame(data)
df
# Делаем выборку компаний, которые относятся к сектору Technology
tech_companies = df[df['Sector'] == 'Technology']
# Рассчитываем долю рыночной стоимости компаний из сектора Technology
tech_market_cap_ratio = tech_companies['Market Cap'].sum() / df['Market Cap'].sum()
# Рассчитываем долю рыночной стоимости компаний всех остальных секторов
other_companies = df[df['Sector'] != 'Technology']
other_market_cap_ratio = other_companies['Market Cap'].sum() / df['Market Cap'].sum()
# Отрисовка диаграммы с сопоставлением результатов
labels = ['Technology', 'Other Sectors']
ratios = [tech_market_cap_ratio, other_market_cap_ratio]
plt.bar(labels, ratios)
plt.xlabel('Sectors')
plt.ylabel('Market Cap (доля)')
plt.title('Market Cap (доля) of Technology vs Other Sectors')
plt.show()
```

```python
# Вставляем список 10 крупнейших компаний компаний
companies = {'AAPL','MSFT','GOOG','GOOGL','AMZN','TSLA','TSM','DXCM','NVDA','MA'}

stock_data = pd.DataFrame()

# Создаем цикл для получения данных по каждой компании
for company in companies:
    ticker = yf.Ticker(company)
    stock_df = ticker.history(period="1d", start="2003-01-01", end="2023-10-01")
    stock_df = stock_df["Close"].pct_change()  # Рассчитываем процентное изменение курса акций
    stock_data[company] = stock_df

# Рассчитываем новый столбец с индексом индустрии на основе 10 крупнейших компаний
index = stock_data.mean(axis=1)
stock_data["Industry Index"] = index

# Информация по акциям компании "Espey Mfg. & Electronics Corp. Common Stock"
espey_ticker = yf.Ticker("ESP")
espey_stock = espey_ticker.history(period="1d", start="2003-01-01", end="2023-10-01")["Close"]

# Построение графиков динамики индекса индустрии и курса акций компании "Espey Mfg. & Electronics Corp. Common Stock"
plt.plot(espey_stock.index, espey_stock, label="Курс акций Espey Mfg. & Electronics Corp. Common Stock")
plt.xlabel("Дата")
plt.ylabel("Курс акций")
plt.title("Курс акций Espey Mfg. & Electronics Corp. Common Stock")
plt.legend()
plt.show()

plt.plot(stock_data.index, stock_data["Industry Index"], label="Индекс индустрии Industrial Machinery/Components")
plt.xlabel("Дата")
plt.ylabel("Индекс Индустрии")
plt.title("Индекс индустрии Industrial Machinery/Components")
plt.legend()
plt.show()
```

```python
# 7. Сравнить динамику показателей мировой экономики (ВВП,
# население, инфляция и т.д., не менее пяти) и динамику индекса.
# Загрузка данных по ВВП
gdp_data = pd.read_csv('gdp.csv')

# Загрузка данных по населению
population_data = pd.read_csv('population.csv')

# Загрузка данных по инфляции
inflation_data = pd.read_csv('inflation-consumer_csv.csv')

# Загрузка данных по индексу потребительских цен
index_data = pd.read_csv('cpi.csv')

# Загрузка данных по Brent Spot Price
gold_price = pd.read_csv('annual.csv')

# Загрузка данных по Natural gas prices
natural_gas = pd.read_csv('monthly_csv.csv')

# Построение графиков для сравнения динамики
plt.figure(figsize=(10, 6))
plt.plot(index_data['Year'], index_data['CPI'], label='Индекс')
plt.xlabel('Год')
plt.ylabel('Значение')
plt.title('Динамика индекса потребительских цен')
plt.legend()
plt.grid(True)
plt.show()

plt.figure(figsize=(10, 6))
plt.plot(gdp_data['Year'], gdp_data['Value'], label='Population')
plt.xlabel('Год')
plt.ylabel('Значение')
plt.title('Динамика населения')
plt.legend()
plt.grid(True)
plt.show()

plt.figure(figsize=(10, 6))
plt.plot(population_data['Year'], population_data['Value'], label='Population')
plt.xlabel('Год')
plt.ylabel('Значение')
plt.title('Динамика населения')
plt.legend()
plt.grid(True)
plt.show()

plt.figure(figsize=(10, 6))
plt.plot(inflation_data['Year'], inflation_data['Inflation'], label='Inflation')
plt.xlabel('Год')
plt.ylabel('Значение')
plt.title('Динамика инфляции')
plt.legend()
plt.grid(True)
plt.show()

plt.figure(figsize=(10, 6))
plt.plot(gold_price['Date'], gold_price['Price'], label='Gold Price')
plt.xlabel('Год')
plt.ylabel('Значение')
plt.title('Динамика Gold Price')
plt.legend()
plt.grid(True)
plt.show()

plt.figure(figsize=(10, 6))
plt.plot(natural_gas['Month'], natural_gas['Price'], label='Natural gas prices')
plt.xlabel('Год')
plt.ylabel('Значение')
plt.title('Динамика Natural gas prices')
plt.legend()
plt.grid(True)
plt.show()
```
