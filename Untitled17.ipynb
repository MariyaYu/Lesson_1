```python
## Ð´Ð¾Ð»Ð¸ Ñ€Ñ‹Ð½Ð¾Ñ‡Ð½Ð¾Ð¹ ÑÑ‚Ð¾Ð¸Ð¼Ð¾ÑÑ‚Ð¸ ÐºÐ¾Ð¼Ð¿Ð°Ð½Ð¸Ð¹ Ð¸Ð· ÑÑ‚Ð¾Ð³Ð¾ ÑÐµÐºÑ‚Ð¾Ñ€Ð°
data = pd.read_excel('Nasdaq.xlsx')
df = pd.DataFrame(data)
df
# Ð”ÐµÐ»Ð°ÐµÐ¼ Ð²Ñ‹Ð±Ð¾Ñ€ÐºÑƒ ÐºÐ¾Ð¼Ð¿Ð°Ð½Ð¸Ð¹, ÐºÐ¾Ñ‚Ð¾Ñ€Ñ‹Ðµ Ð¾Ñ‚Ð½Ð¾ÑÑÑ‚ÑÑ Ðº ÑÐµÐºÑ‚Ð¾Ñ€Ñƒ Technology
tech_companies = df[df['Sector'] == 'Technology']
# Ð Ð°ÑÑÑ‡Ð¸Ñ‚Ñ‹Ð²Ð°ÐµÐ¼ Ð´Ð¾Ð»ÑŽ Ñ€Ñ‹Ð½Ð¾Ñ‡Ð½Ð¾Ð¹ ÑÑ‚Ð¾Ð¸Ð¼Ð¾ÑÑ‚Ð¸ ÐºÐ¾Ð¼Ð¿Ð°Ð½Ð¸Ð¹ Ð¸Ð· ÑÐµÐºÑ‚Ð¾Ñ€Ð° Technology
tech_market_cap_ratio = tech_companies['Market Cap'].sum() / df['Market Cap'].sum()
# Ð Ð°ÑÑÑ‡Ð¸Ñ‚Ñ‹Ð²Ð°ÐµÐ¼ Ð´Ð¾Ð»ÑŽ Ñ€Ñ‹Ð½Ð¾Ñ‡Ð½Ð¾Ð¹ ÑÑ‚Ð¾Ð¸Ð¼Ð¾ÑÑ‚Ð¸ ÐºÐ¾Ð¼Ð¿Ð°Ð½Ð¸Ð¹ Ð²ÑÐµÑ… Ð¾ÑÑ‚Ð°Ð»ÑŒÐ½Ñ‹Ñ… ÑÐµÐºÑ‚Ð¾Ñ€Ð¾Ð²
other_companies = df[df['Sector'] != 'Technology']
other_market_cap_ratio = other_companies['Market Cap'].sum() / df['Market Cap'].sum()
# ÐžÑ‚Ñ€Ð¸ÑÐ¾Ð²ÐºÐ° Ð´Ð¸Ð°Ð³Ñ€Ð°Ð¼Ð¼Ñ‹ Ñ ÑÐ¾Ð¿Ð¾ÑÑ‚Ð°Ð²Ð»ÐµÐ½Ð¸ÐµÐ¼ Ñ€ÐµÐ·ÑƒÐ»ÑŒÑ‚Ð°Ñ‚Ð¾Ð²
labels = ['Technology', 'Other Sectors']
ratios = [tech_market_cap_ratio, other_market_cap_ratio]
plt.bar(labels, ratios)
plt.xlabel('Sectors')
plt.ylabel('Market Cap (Ð´Ð¾Ð»Ñ)')
plt.title('Market Cap (Ð´Ð¾Ð»Ñ) of Technology vs Other Sectors')
plt.show()
```
